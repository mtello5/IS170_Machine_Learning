# **Naive Bayes **

The kids can play outside if the outlook and temp permit. Any adversarial outlook or high or low temp, the kids can't.

```python
import pandas as pd
import numpy as np

df=pd.read_csv("/content/player.csv")
df
```

```python
#In order to process we must turn those categorical values into numberic values
#Import LabelEncoder
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

#Using .fit_transform function to fit label
#encoder and return encoded label
outlook_label=le.fit_transform(df['outlook'])

#printing outlook_label
outlook_label
```

```python
#repeat it for temp label
temp_label=le.fit_transform(df['temperature'])

#printing temp_label
temp_label
```

```python
#repeat it for play label
play_label=le.fit_transform(df['playgolf'])

#printing play_label
play_label
```

```python
#we'll use zip method in this function called merge to combine the two labels: outlook and temp
def merge (list1, list2):
  merged_list=tuple(zip(list1, list2))
  return merged_list

features=merge(outlook_label, temp_label)
features
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

# Train the model using the training sets
model.fit(features,play_label)

#Predict Output
predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
print ("Predicted Value:", predicted)
```

Based if outlook is overcast (0) and temperature is mild (2) it is predicted to play golf

# **Naive Bayes using Multiple Labels**

```python
df2=pd.read_csv("/content/player.csv")
df2.head()
```

```python
categ = ['outlook','temperature','humidity','wind','playgolf']

# Encode Categorical Columns
le = preprocessing.LabelEncoder()
df2 = df2[categ] = df2[categ].apply(le.fit_transform)
df2
```

```python
x_data = df2[['outlook', 'temperature', 'humidity', 'wind']]
y_data = df2['playgolf']
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.3,random_state=109) # 70% training and 30% test
```

```python
#This is model generation

#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB

#Create a Gaussian Classifier
model = GaussianNB()

#Train the model using the training sets
model.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)

#This is to evaluate the model
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

With this model, we received a 40% accuracy in the factors of outlook, temperature, humidity, and wind in determing if we will be able to play golf.
