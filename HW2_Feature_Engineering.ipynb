```python
df  = pd.read_csv('list-cities-california-state-us-168j.csv')
print(df)
```

```python
# 1 Deal with missing or null values
df[['FullStreetName', 'StreetName', 'StreetType', 'PostDirection']].dropna()
```

```python
# String to integer conversion
mapping = {'ST': 1, 'AVE': 2, 'RD': 3, 'CT': 4, 'BLVD': 5, 'WAY': 6, 'CT': 7}
df['StreetType'].fillna(-1, inplace=True)
df['StreetType'] = df['StreetType'].map(mapping).fillna(-1).astype(int)
print(df)
```

```python
df2  = pd.read_csv('2010_Census_Populations_by_Zip_Code.csv')
print(df2)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

integer_cols = ['Total Population', 'Median Age', 'Total Males', 'Total Females']

df2[integer_cols] = scaler.fit_transform(df2[integer_cols])
print(df2)
```

```python
#4 Column Manipulation

# Renaming columns
df2 = df2[['Total Population', 'Median Age']]

hc = df2.rename(columns={'Total Population': 'Population of ZIP',
                                  'Median Age': 'Average Age'},
                         inplace=False)
hc
```
