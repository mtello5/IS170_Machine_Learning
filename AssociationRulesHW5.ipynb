```python
#set up the libraries
# we use apriori for SUPPORT and association_rules for CONFIDENCE, LIFT and CONVICTION
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
#Loading the dataset
dataset=pd.read_csv('/content/G1list.csv')
dataset.head()
```

```python
#Since it's missing a column header, so we create new column header name
df=pd.read_csv('/content/G1list.csv', sep=',')
df.head()
```

```python
#data size
df.shape
```

```python
#we are turning the pandas dataframe into a list of lists
data=list(df["Groceries"].apply(lambda x:x.split(",")))
data
```

```python
#Let's transform the list with one-hot encoding
a=TransactionEncoder()
a_data=a.fit(data).transform(data)
df=pd.DataFrame(a_data, columns=a.columns_)
df=df.replace(True,1)
df=df.replace(False,0)
df
```

```python
#Set a threshold value for the support value and calculate the support value
df=apriori(df,min_support=0.2, use_colnames=True)
df
```

SUPPORT tells us how frequent each of the items are appearing in our dataset. In this example, we know bananas appears the most with SUPPORT measure of .31. Further, we know bread is the most likely to be seen in a transaction with SUPPORT value .44

```python
#Here is the CONFIDENCE (Try different min_threshold value to explore this more)
df_ar=association_rules(df,metric="confidence", min_threshold=0.3)
df_ar
```

We can see (Coffee -> Bread) show CONFIDENCE 0.6 which means the two items are bought together 60% of the time. Additionally, we can see (Bread -> Coffee) shows CONFIDENCE 0.46 meaning when someone buys Bread 46% of the time they buy coffee with it together.

```python
#Here is LIFT (try different min_threshold value to explore more) 
df_ar=association_rules(df,metric="lift", min_threshold=1)
df_ar
```

The first row (coffee & bread) show LIFT 1.33 which is only slight positive correlation between the two. The next row, (Bread -> Coffee) show LIFT 1.33 which is the same because A->C & C->A there is no antecedent or consequent. 

```python
#Let's view our interpretation values using the Association rule function
df_ar=association_rules(df,metric="conviction", min_threshold=0.6)
df_ar
```

CONVICTION is same as LIFT except CONVICTION also gives the directed measure. The first row (Coffee -> Bread) CONVICTION shows 1.37 which is pretty strong relationship, but the second row (Bread -> Coffee) CONVICTION is 1.21 which is also strong relationship but not as strong. The higher than 1 the CONVICTION value the higher the relationship. The antecedent in this case Coffee has a stronger relationship to Bread than Bread to Coffee.

```python
#Loading the dataset
dataset3=pd.read_csv('/content/BreakfastData.csv')
dataset3.head()
```

```python
#Since it's missing a column header, so we create new column header name
df3=pd.read_csv('/content/BreakfastData.csv', names=['products'])
df3.head() 
```

```python
#data size
df3.shape
```

```python
#we are turning the pandas dataframe into a list of lists
data3=list(df3["products"].apply(lambda x:x.split(",")))
data3
```

```python
#Let's transform the list with one-hot encoding (make sure you understand the one-hot encoding concept first)
#Each column represents an item and each row represent a record or a transaction. 
#Initially, if that particular item occurs then it's labeled TRUE, if not then FALSE.
#After that, TRUE is converted to 1 and FALSE to 0 because computer loves numeric values.

from mlxtend.preprocessing import TransactionEncoder
a=TransactionEncoder()
a_data=a.fit(data3).transform(data3)
df3=pd.DataFrame(a_data, columns=a.columns_)

#Turn False & True to 0 & 1 for easy readability
df3=df3.replace(False,0)
df3=df3.replace(True,1)

df3
```

```python
#We'll try 0.2 for the SUPPORT threshold value (but you can set it to any value you want).
# The parameter (use_colnames=True) is used to convert the integer back to its respective name.

df3=apriori(df3,min_support=0.2,use_colnames=True)
df3
```

Bacon we can see has SUPPORT value .545 which if we did it manually means it appears in 5/11 transactions and overall was most prominent in over half. Additionally we can see some itemsets with more than one item also showing SUPPORT. On the other side, we can see lower SUPPORT level mean they itemset is less likely to appear in the transaction.

```python
#Let's view our interpretation values using the Association rule function.
df3_ar=association_rules(df3,metric="lift", min_threshold=1)
df3_ar
```

Using LIFT to determine values of relationship we can see relationship of items bought seperately from each other. We can see that they have a postive correlation with LIFT value of 1.222. We can also see eggs and sausage with a LIFT value of 1.375 will a stronger postive correlation.

```python
#Let's view our interpretation values using the Association rule function.
df3_ar=association_rules(df3,metric="confidence", min_threshold=0.6)
df3_ar
```

CONFIDENCE is used to tell us probability the antecedant is purchased in the same transcation as the consequent. In this case, if a customer buys eggs there is a 66.7% chance they buy bacon also. Additionally, If someone buys sausage with CONFIDENCE value of .75 there is a 75% chance they buy eggs with it also.

```python
#Let's view our interpretation values using the Association rule function
df3_ar=association_rules(df3,metric="conviction", min_threshold=0.6)
df3_ar
```

Lastly, CONVICTION lets us know how probable it A appears without C. So by analyzing our list of CONVICTION values we see (Eggs->Bacon) and (Sausage->Eggs) have values above 1 meaning they have a strong relationship. All  the other values we see in the table show CONVICTION values >1 meaning they have negative relationships. This means they antecedent makes the consequent less likely to be in the transaction (coffee -> bacon)
